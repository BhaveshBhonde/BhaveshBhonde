About Dataset:
The data is about of pune water waterbodies to find out if it is safely able use it or not.This is important for regional development for Pune.

Content
Temperature:
Temperatures are around 20 degrees Celsius (68 degrees Fahrenheit), up to 30 degrees Celsius (86 degrees Fahrenheit).

pH value:
pH is an important parameter in evaluating th acid-base balance of water.It is also the indicator of acidic or alkaline conditionof water status.its limit is from 6.5 to 8.5.

Electric Conductivity:
Pure water is not a good conductor of electric current rather's a good insulator.Increase in ions concentration enhances the electrical conductivity of water.Generally the amount of dissolved soilds in water dtermines the electric conductivity (EC) actually measure the ionic process of a solution that enables it to transmit current.

Total Hardness:
Hardness is mainly caused by calcium andmagnesium salts. These salts are dissloved from geologic deposits through which water travels.The length of time water is in contact with hardness producing material helps determine how much hardness there is in raw water.Hardness was originally defined as the capacity of water to predict soap caused by Calcium and Magnesium.

Dissolved Oxygen (mg/L):
Dissolved oxygen (DO) is the amount of oxygen gas dissolved in water.DO is a critical indicator of water quality. By monitoring DO levels, we can help to protect aquatic life and ensure that our water resources are healthy.

Biochemical Oxygen Demand (BOD) (mg/L):
Biochemical oxygen demand (BOD) is an important measure of water pollution. It is the amount of dissolved oxygen (DO) that is needed by aerobic microorganisms to break down organic matter in water over a specific time period, typically five days. BOD is measured in milligrams of oxygen per liter of water (mg/L).

Total Dissolved Solids (TDS) (mg/L):
Total dissolved solids (TDS) is a measure of the amount of dissolved inorganic and organic substances in water. These substances can include minerals, salts, metals, and ions. TDS is typically measured in parts per million (ppm) or milligrams per liter (mg/L).

Fecal Coliform (CFU/100 mL):
Fecal coliform bacteria are a group of bacteria that are present in the feces of warm-blooded animals, including humans. They are a type of indicator bacteria, which means that their presence indicates that water may be contaminated with fecal matter. Fecal coliform bacteria are not harmful in themselves, but they can be an indicator of the presence of other harmful pathogens, such as viruses and parasites.

Potability:
Indicates if water is safe for Human use where 1 means Potable and 0 means Not Potable.(0)Water is not safe to use and(1)Water is safe to use

# Train Decision Tree Classifier and Check accuracy

```python
data= pd.read_csv(r"C:\Users\HP\Project Work\water quality data.csv")
```

```python
X= data.drop('Potability',axis=1 )# Input Variable
Y= data['Potability'] # Target Variable
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, shuffle=True, random_state=None)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
dt=DecisionTreeClassifier(criterion='gini',min_samples_split=5, splitter='best')
dt.fit(X_train,Y_train)
```

```python
prediction=dt.predict(X_test)
print(f"Accuracy Score={accuracy_score(Y_test,prediction)*100}")
print(f"Confusion Matrix=\n{confusion_matrix(Y_test,prediction)}")
print(f"Classification Report=\n{classification_report(Y_test,prediction)}")
```

# Apply Hyper Parameter Tunning

```python
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV

# define models and parameters
model=DecisionTreeClassifier()
criterion=['gini','entropy']
splitter=['best','random']
min_samples_split=[2,4,6,8,10,12,14]

#define grid search
grid = dict(splitter=splitter, criterion=criterion, min_samples_split=min_samples_split)
cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=3, random_state=1)
grid_search_dt = GridSearchCV(estimator=model, param_grid=grid, n_jobs=-1, cv=cv, scoring='accuracy', error_score=0)
grid_search_dt.fit(X_train,Y_train)
```

```python
print(f"Best:  {grid_search_dt.best_score_:.3f} using {grid_search_dt.best_params_}")
means = grid_search_dt.cv_results_['mean_test_score']
stds = grid_search_dt.cv_results_['std_test_score']
params = grid_search_dt.cv_results_['params']

for mean, stdev, param in zip(means,stds,params):
    print(f"{mean:.3f} ({stdev:.3f}) with:{param}")
    
print("Training Score:",grid_search_dt.score(X_train,Y_train)*100)
print("Testing Score:", grid_search_dt.score(X_test, Y_test)*100)
```
